```python
#p[0].df().loc[:, ['GR','DT','RHOB','CALI','ILD']]
#n = pd.DataFrame(['wellName','depth','GR','DT','RHOB','CALI','ILD', 'Top'])
n = pd.DataFrame(columns=['wellname','depth','GR','DT','RHOB','CALI','ILD', 'Top','TopX','TopY'])
```

```python
well_tops = pd.read_csv('oga_markers_selected_wells_initial.csv')
well_by_las = pd.read_csv('wells_by_lasfile_edit.csv')
well_tops.drop_duplicates(subset='Well_Number',keep='first', inplace=True)
well_names = well_tops
```

```python
well_tops['Well_Number'] = well_tops['Well_Number'].str.replace(' ','').str.lower()
well_by_las['wname'] = well_by_las['wname'].str.replace(' ','').str.lower()
```

```python
well_tops
```

```python
for i, row in well_by_las.iterrows():
    #print(row["fileName"])
    #print(type(row["fileName"]))
    fileName = row["fileName"]
    print(fileName)
    well_log = Project.from_las(fileName)
    my_df = well_log[0].df()
    my_df = my_df.assign(wellname = well_log[0].header.name)
    my_df = my_df.assign(Top = 'NaN')
    my_df = my_df.assign(TopX = 'NaN')
    my_df = my_df.assign(TopY = 'NaN')
    #n.concat(my_df, join = 'inner', axis = 0, inplace=True)
    n = pd.concat([n, my_df], join = 'inner', axis = 0)
    #print(pd.concat([n, my_df], join = 'inner', axis = 0).head())
    print(n.tail())
    print(len(n))
```
