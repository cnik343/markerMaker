# Making Maps with Geopandas

GeoPandas is an open source project to make working with geospatial data in python easier. GeoPandas extends the datatypes used by pandas to allow spatial operations on geometric types - see http://geopandas.org/index.html for full details. Important features of GeoPandas are the ability to load standard GIS shapefiles, and the reprojection of geodata frames into new Coordinate Reference Systems (CRS). This is demonstrated here by generating some example plots from test data and then for the UKCS and Norwegian quad data, plotting it with some European coastlines.

First install the libraries, some GIS data and adjust some of the notebook properties.

```python
# Import the libraries
import re
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
import matplotlib.pyplot as plt
from os.path import join


# Locate the different GISdata directories
neDir = '../../#gisData/naturalEarth-10m'
ogaWGS84 = '../../#gisData/oga-offshore-wgs84'
ogaED50 = '../../#gisData/oga-offshore-ed50'
npdED50 = '../../#gisData/npd-offshore-ed50'
```

### Import the well data...

Load the full and partial marker spreadsheets from file. In each case ensure that the location is defined as `X, Y`, and use this to generate the `XY` geometry required to turn the pandas dataframe into a geopandas dataframe. Lastly define a function to perform this conversion for a specified reference geometry and output coordinate system - EPSG code 23090. NB: Projections can be defined in a variety of ways by referencing the ellipsoid, datum and offsets etc, but common projections can also be referred to by their EPSG codes (see http://spatialreference.org/).

```python
# Import the well data and turn them into a geopandas dataframe...
markersRaw = pd.read_csv('../data/all_well_tops.csv')
markersPal = pd.read_csv('../data/pal_well_tops.csv')
markersPal.rename(columns={"x": "X", "y": "Y"}, inplace=True)

# Use a combination of Regex and concat to restore the XY locations for the DF with all the markers
pattern = '\d+\.\d+'
markerXY = pd.DataFrame([re.findall(pattern, item) for item in markersRaw['XY']], columns=['X','Y'])
markersAll = pd.concat([markersRaw, markerXY], axis=1).drop(['Unnamed: 0', 'XY'], axis=1)
markersAll['X'] = pd.to_numeric(markersAll['X'])
markersAll['Y'] = pd.to_numeric(markersAll['Y'])


def gpd_convert(markers):
    
    '''Convert the XY location of a marker spreadsheet into a geopandas geometry'''
    
    markers['XY'] = list(zip(markers['X'], markers['Y']))
    markers['XY'] = markers['XY'].apply(Point)
    geomarkers = gpd.GeoDataFrame(markers, geometry='XY')
    geomarkers.crs = {'init' :'epsg:4326'}
    geomarkers = geomarkers.to_crs({'init': 'epsg:23090'})

    return geomarkers

geomarkersAll = gpd_convert(markersAll)
geomarkersPal = gpd_convert(markersPal)
```

Define a function to extract the first marker for each well, as a proxy for the well location, and apply this to both the marker dataframes in order to plot well locations. Make sure the geopandas geometry is set.

```python
def wells_locs(geomarkers):
    
    '''Extract the XY location of the first marker for each well into a geopandas geometry'''
    
    wells = geomarkers[['XY', 'Well_Number']].groupby('Well_Number').first().reset_index()
    geowells = gpd.GeoDataFrame(wells, geometry='XY')
    
    return geowells

wellsAll = wells_locs(geomarkersAll)
wellsPal = wells_locs(geomarkersPal)
```

### Using geopandas to load and plot GIS shapefiles

Geopandas can be used to directly load GIS shapefiles, which makes it a good tool for visualising oil and gas data spatially in the python environment. Load the UKCS and Norwegian Quad shapefiles and plot them with a high resolution version of the Natural Earth countries shapefile. Prior to plotting reproject the geopandas dataframes onto a new CRS using the *gpd.to_crs()* method with EPSG 23090, which is a ED50 Transverse Mercator projection using ED50 centred on the Greenwich Meridian, designed to display data across the North Sea, between about -3 -- +3 degrees of longitude.

For Reference, the Natural Earth database can be found here: http://www.naturalearthdata.com/downloads/

```python
# Reload the OGA quads shapefile
ogaQuadsWGS84 = gpd.read_file(join(ogaWGS84, 'OGA_Quadrants_WGS84.shp'))
ogaQuadsED50 = gpd.read_file(join(ogaED50, 'OGA_Quadrants_ED50.shp'))

# Reload the NPD quads shapefile
npdQuadsED50 = gpd.read_file(join(npdED50, 'v_geo_quadrant.shp'))

# Reload and slice the Hi-Res Natural Earth countries shapefile
worldHR = gpd.read_file(join(neDir, 'ne_10m_admin_0_countries.shp'))
worldHR = worldHR[(worldHR.CONTINENT == 'Europe')]

# Change the CRS to TM0 using the gpd.to_crs() method (Note, gpd.to_crs(epsg=23090) would also work)
ogaQuadsWGS84 = ogaQuadsWGS84.to_crs({'init': 'epsg:23090'})
ogaQuadsED50 = ogaQuadsED50.to_crs({'init': 'epsg:23090'})
npdQuadsED50 = npdQuadsED50.to_crs({'init': 'epsg:23090'})
worldHR = worldHR.to_crs({'init': 'epsg:23090'})
```

### Plot the well locations

Now plot the wells and country data in the desired transverse mercator projection...

```python
# Generate the TM0 plot
fig, ax = plt.subplots(figsize=(18, 12))
ax.set_aspect('equal')

ogaQuadsED50.plot(ax=ax, linewidth=0.4, alpha=0.9, color='lightblue', edgecolor='black')
npdQuadsED50.plot(ax=ax, linewidth=0.4, alpha=0.9, color='lightblue', edgecolor='red')
worldHR.plot(ax=ax)

wellsAll.plot(ax=ax, markersize=4, color='c')
wellsPal.plot(ax=ax, markersize=4, color='r')

plt.xlim([220000, 880000])
plt.ylim([6120000, 6940000])
plt.show();
```
