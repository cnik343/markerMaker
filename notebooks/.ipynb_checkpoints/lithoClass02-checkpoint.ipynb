```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from xgboost.sklearn import XGBClassifier
from fancyimpute import IterativeImputer
from sklearn import svm
```

```python
# Import the log data...
# df = pd.read_csv('../data/dummy_well.csv')
df = pd.read_csv('../data/wells_processed_04.csv')
df.Top = df.Top.fillna(value='Unknown')
print(df.isna().sum())
print(df.tail())
```

```python
df_known = df[df['Top'] != 'Unknown']
print(df_known.tail())

# predNames = ['CALI', 'GR', 'DT', 'RHOB', 'ILD', 'NPHI']
predNames = ['GR', 'DT', 'RHOB', 'ILD']

predDF = df_known[predNames]
df.Top.unique()
```

```python
imputer = IterativeImputer(n_iter=5, random_state=42)
predDFfull = pd.DataFrame(imputer.fit_transform(predDF),  columns=predNames)
    
# predDFfull = df.fillna(value=float('-99999'))
print(predDFfull.isna().sum())
print(predDFfull.tail())
```

```python
scaler = StandardScaler().fit(predDFfull)
```

```python
scaledPredictors = pd.DataFrame(scaler.transform(predDFfull),
                                columns=predNames)
```

```python
scaledPredictors.head()
scaledPredictors.info()
```

```python
X_train, X_test, y_train, y_test = train_test_split(scaledPredictors,
                                                    df_known.Top,
                                                    test_size=0.3,
                                                    random_state=42)
```

```python
X_test.head()
```

```python
# svmClass = svm.SVC(kernel='linear', C=1)
# svmClass.fit(X_train, y_train)
# svmClass.score(X_test, y_test)       
```

```python
xgbClass = XGBClassifier()
xgbClass.fit(X_train, y_train)
xgbClass.score(X_test, y_test) 
```

```python
# make predictions for test data
y_pred = xgbClass.predict(X_test)

# evaluate predictions
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```

```python
print(classification_report(y_test, xgbClass.predict(X_test), digits=3))
```

```python
# itertoools is a standard library for all kinds of handy iterator manipulation
import itertools

# Compute confusion matrix
cnf_matrix = confusion_matrix(y_test, y_pred)
selected = np.unique(y_test)

title = 'Confusion matrix'
cmap = plt.cm.Reds

# Plot non-normalized confusion matrix.
fig, ax = plt.subplots(figsize=(18, 18))
plt.imshow(cnf_matrix, interpolation='nearest', cmap=cmap)
plt.title(title)
# plt.colorbar()
tick_marks = np.arange(len(selected))
plt.xticks(tick_marks, selected, rotation=45)
plt.yticks(tick_marks, selected)

# Print the support numbers inside the plot.
thresh = cnf_matrix.max() / 2.
for i, j in itertools.product(range(cnf_matrix.shape[0]), range(cnf_matrix.shape[1])):
    plt.text(j, i, format(cnf_matrix[i, j], 'd'),
             horizontalalignment="center",
             color="white" if cnf_matrix[i, j] > thresh else "black")

plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.tight_layout()
# plt.show()
plt.savefig('../figures/con_matrix.png')
```
